```python
import math
again = True
while again:
        choice = input("1 for Adding: , 2 for Subracting:  , 3 for multiplying: , 4 for division: , 5 for percentage: , 6 for squareroot:")

        if choice in ('1', '2', '3', '4', '5', '6'):
            try:
                x = int(input("Enter first number: "))
                y = int(input("Enter second number: "))
            except:
                print("Invalid")
            finally:
                print("")

            if choice == '1':
                print(f"x + y = {sum(x , y)}")

            elif choice == '2':
                print(f"x - y = {x - y}")

            elif choice == '3':
                print(f"x * y = {math.prod(x, y)}")

            elif choice == '4':
                print(f"x // y = {x // y}")

            elif choice == '5':
                print(f"(x / y)*100 = {(x // y)*100}")

            elif choice == '6':
                print(f"math.sqrt(x or y) = {math.sqrt(x)}")
        else:
            print("Invalid Input.")
        again = bool(int(input("Do you want to stop : ")))
```
